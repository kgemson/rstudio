### Task 6: Visualisation activity (12 marks)

````python
# Task 6a answer here:

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Assign 'hdi_by_year.csv' to a variable and read_csv to read this into a dataframe
mydatafile2 = "d:\\Users\\kgems\\Documents\\NPA\\Python\\hdi_by_year.csv"
df2 = pd.read_csv(mydatafile2)

# Filter out just the selected country
iceland_df = df2[df2['Country'] == ' Iceland']

# Now transpose as easier to plot (and rename column)
iceland_df_t = iceland_df.transpose()
iceland_df_t.columns = ['HDI']

# Assign variables from dataframe, ignoring first two rows of index and values
years = iceland_df_t.index[2:]
hdi = iceland_df_t['HDI'].iloc[2:]

# Now plot values
plt.figure(figsize=(20,10))
plt.plot(years, hdi)
plt.title('HDI score per year for Iceland', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('HDI', fontsize=14)
plt.show()
````

````python
# Task 6b bar chart - GNI per capita bar chart for three countries
# Import libraries
#import pandas as pd
#import matplotlib.pyplot as plt

# Assign 'hdi_by_year.csv' to a variable and use 'read_csv' to read this into a dataframe
mydatafile = "d:\\Users\\kgems\\Documents\\NPA\\Python\\hdi_and_components.csv"
df = pd.read_csv(mydatafile)

# Filter out just the selected countries
country_list = ['Iceland','Spain','Italy']
comparison_df = df[df['Country'].isin(country_list)][['Country','Life expectancy at birth (years)', 'Expected years of schooling (years) 2019', 'Gross national income (GNI) per capita (2019)']]
comparison_df

# Now plot values
plt.figure(figsize=(20,10))
plt.bar(comparison_df['Country'], comparison_df['Gross national income (GNI) per capita (2019)'], color=['green', 'red', 'blue'])
plt.title('Gross national income (GNI) per capita (2019)', fontsize=18)
plt.xlabel('Country', fontsize=16)
plt.ylabel('GNI per capita', fontsize=14)
plt.show()
````
