### Description of the pizza dataset
A synthetic dataset that describes pizza sales for a pizza place somewhere in the US. While the contents are artificial, the ingredients used to make the pizzas are far from it. There are 32 different pizzas that fall into 4 different categories: classic (classic pizzas: 'You probably had one like it before, but never like this!'), chicken (pizzas with chicken as a major ingredient: 'Try the Southwest Chicken Pizza! You'll love it!'), supreme (pizzas that try a little harder: 'My Soppressata pizza uses only the finest salami from my personal salumist!'), and, veggie (pizzas without any meats whatsoever: 'My Five Cheese pizza has so many cheeses, I can only offer it in Large Size!').

### Fields in the dataset
* id - The ID for the order, which consists of one or more pizzas at a give date and time 
* date - A character representation of the order date, expressed in the ISO 8601 date format (YYYY-MM-DD)
* time - A character representation of the order time, expressed as a 24-hour time the ISO 8601 extended time format (hh:mm:ss)
* name - The short name for the pizza
* size - The size of the pizza, which can either be S, M, L, XL (rare!), or XXL (even rarer!); most pizzas are available in the S, M, and L sizes but exceptions apply
* type - The category or type of pizza, which can either be classic, chicken, supreme, or veggie
* price - The price of the pizza and the amount that it sold for (in USD)

Q1 Import the ```pizza.csv``` file using pandas and create a ```df``` datframe variable, print out the whole ```df```

````python
# your code here
# import pandas and alias as pd
import pandas as pd
import matplotlib.pyplot as plt

# create the data frame (df) variable and then print
datafile = "d:\\Users\\kgems\\Documents\\NPA\\Python\\pizza.csv"

# Tell python to open the file as a variable
df = pd.read_csv(datafile)
    
# Loop over the file and print out all the rows
print(df)
````

Q2 Add the code to generate a bar chart showing the most common type of pizza.

````python
# answer below this line

# Group dataframe elements by type
type_df = df.groupby('type')['id'].count()
print(type_df)

# Print bar chart using counts of pizza types
type_df.plot.bar(figsize=(10,6), rot=1)
plt.show()
````

Q3 Add a main title to the barplot, add an x and y label to the barplot, colour the bars in the bar chart with 4 different colours.

````python
# answer below this line
# Set display parameters
colors_list = ['gold', 'red', 'green', 'blue']

type_df.plot.bar(color=colors_list, figsize=(10,6), rot=1)
plt.title('Count of pizza sales by type', fontsize=18)
plt.xlabel('Pizza type', fontsize=14)
plt.ylabel('Count')
plt.show()
````

Q4 Create a barchart of the most common pizza sizes, add a main title to the barplot, add an x and y label to the barplot

````python
# answer below this line
# Group dataframe elements by size
size_df = df.groupby('size')['id'].count()

# Print bar chart using counts of pizza sizes
colors_list = ['salmon', 'orange', 'lightgreen', 'lightblue', 'silver']
size_df.plot.bar(color=colors_list,figsize=(10,6),rot=1)
plt.title('Count of pizza sales by size', fontsize=18)
plt.xlabel('Pizza size', fontsize=14)
plt.ylabel('Count')
plt.show()
````

Q5 Calculate the mean price of a pizza with pandas.

````python
# answer below this line
print(round(df['price'].mean(),2))
````

Q6 Calculate the Min and Max pizza prices using pandas.

````python
# answer below this line
print('Min price:', round(df['price'].min(),2), 'Max price:', round(df['price'].max(),2))
````

Q7 Create a bar chart showing the numer of sales of the most popular veggie pizzas, do the same for classic pizzas.

````python
# answer below this line
# Create dataframes containing only veggie / classic pizzas
veggie_df = df[df['type'] == 'veggie']
classic_df = df[df['type'] == 'classic']

# Group veggie / classic pizzas by name
veggie_name_df = veggie_df.groupby('name')['id'].count()
classic_name_df = classic_df.groupby('name')['id'].count()

# Print bar chart using counts of pizza names
colors_list = ['aqua', 'lightblue', 'gold', 'red', 'green', 'blue', 'purple', 'orange', 'salmon']

veggie_name_df.plot.bar(color=colors_list,figsize=(12,6),rot=1)
plt.title('Count of veggie pizza sales', fontsize=18)
plt.xlabel('Veggie pizza name', fontsize=14)
plt.ylabel('Count')
plt.show()

classic_name_df.plot.bar(color=colors_list,figsize=(12,6),rot=1)
plt.title('Count of classic pizza sales', fontsize=18)
plt.xlabel('Classic pizza name', fontsize=14)
plt.ylabel('Count')
plt.show()
````
