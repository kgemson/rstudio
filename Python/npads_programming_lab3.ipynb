Q3.8 Create pie chart showing the split of different political parties, each slice of the pie chart should be a different color. 
To do this you will need to make a total of each political party in the file, then place these in a pie chart.

````python
# your code here
# import pandas and alias as pd, matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt

# create the data frame (df) variable
datafile = "d:\\Users\\kgems\\Documents\\NPA\\Python\\uk_prime_ministers.csv"

# Tell python to open the file as a variable
df = pd.read_csv(datafile)
    
# Group dataframe elements by party
grouped_df = df.groupby(['party']).sum()
grouped_df = grouped_df.sort_values(by='length_of_term')

# Show data as pie chart
# Set display parameters
colors_list = ['green', 'red', 'gold', 'lightblue', 'orange', 'blue']

fig = plt.figure(figsize=(10,10))
fig.patch.set_facecolor('lightgreen')
fig.patch.set_alpha(0.7)

plt.axis('equal')
plt.title('Prime Ministerial length of term by political party', fontsize=18)

# Set up pie chart parameters and display
plt.pie(grouped_df['length_of_term'],labels=grouped_df.index, colors=colors_list,autopct='%1.1f%%',shadow=True,textprops={'fontsize': 16})

plt.show()
````
